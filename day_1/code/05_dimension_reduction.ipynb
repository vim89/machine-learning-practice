## Dimension Reduction Naive Technique

Let's just remove features?!

We can reduce the number of features by either:
1. Manually reducing the number of features included in the model

We can manually remove variables of low signifiance, low correlation, variables with collinearity and low importance. Variables such as name are unlikely to be useful!

2. Introduce bias into the model (LASSO)

LASSO (least absolute shrinkage and selection operator) and is a bias that penalizes variables in a model!
3. Automatically reducing the number of features in the model with a selection technique (backwords, stepwise, etc.)

Remove variables one at a time with a stopping mechanism

### Principal Components Analysis

PCA is an unsupervised technique that we can use on continous data. The idea is that we want to generate principal components that represent a high amount of the variation in our data. The below example shows how we use PCA to 



The math behind PCA uses the mean and covariance to determine the direction of highest variance.
1. Compute the mean of every feature
2. Compute a covariance matrix

The formula is:

$ Cov(X,Y) = \sum \frac{(X-μ)(Y-ν)}{n-1} $ where:

μ is the the mean of X and

ν is the the mean of Y

n = the number of items in the data set



3. Compute the eigenvectors and eigenvalues
eigenvectors and eigenvalues represent the direction with highest variance, and the number of eigenvalues is the number of features.

Let's get into some math!
Mathematically, and eigenvalue is calculated by

$\lvert A-\lambda * i \rvert$ = 0

Where A is a matrix, lambda is our eigenvalue, and I is an identity matrix (matrix with just ones along the diagonals)

The determinant of a matrix is
|A| = ad − bc

4. Sort eigenvalues by decreasing value and select k eigenvalues (explaining an amount of variation we choose)

5. Transform features into k dimesions, 1 for each eigenvalue
