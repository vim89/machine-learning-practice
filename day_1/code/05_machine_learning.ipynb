### What do you think Machine Learning is?  How can a machine learn?
Machine Learning is a broad field that sits at the intersection of math / statistics and computer science. 

**The general process we will discuss with machine learning is:**
1. A algorithm structure - how does the algorithm work and what are the components of the algorithm?

For instance, how does a decision tree work?  How does a linear regression work?
2. An evaluation criteria - how do we know if our machine learning algorithm is working efficiently and how will we judge?

For example, in a binary classification, we look at accuracy, misclassification, in regression we might consider MAPE (mean absolute percent error)
3. An optimization method - how do we determine the mathematical coefficients of our algorithm to generate the best evaluation criteria!

For example, linear regression uses least squares



### Types of Machine Learning

In ML we have three different types of algorithms

1. Supervised: The model has a target variable- a variable we are interested in understanding and predicting

2. Unsupervised: The model has no target variable- we are interested in understanding the data

3. Reinforcement Learning: We are building an agent designed to exploit and understand it's environment with a reward system.

Also - Semi-supervised exists, where we have some instances of a label (small number), and wish to learn more about this class.
