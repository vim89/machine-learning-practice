# Feature Engineering Exercise

## In this exercise we will explore different feature engineering ideas, our process will follow below
1. Explore Datasets and answer the important data questions 

    What is the format of the data?
    
    How much data do we have?
    
    Describe the data?
    
    Where is the data from?
    
    What is the problem we are trying to solve?
    
    **What features might be key in solving this problem?**
    
2. Form hypothesis and gather additional data if necessary, conduct interviews and gain domain knowledge

3. Experiment with potential features and understand correlations

4. Discuss features with audience and gain further domain knowledge

5. Build models and refine features as necessary
