# Feature Engineering


### What is feature engineering?

Feature engineering is the process of generating an input vector for machine learning, in a traditional sense, we want to have a single vector of data per observation! Feature engineering is more than just data cleaning though, it is the process of using our **understanding of the problem and domain knowledge to create datapoints more directly capable of generating an accurate model. **

Kaggle competition winners have often ranged from two groups
A. Models made with extremely deep neural networks and complex ensemble solutions
B. Models made with deep understanding of the problem and detailed / precise feature engineering

### Categorical Data

In categorical data we can start off with One-Hot encoding and label encoding. Understanding if we have ordinal or non-ordered data is our first step. From here we can generate interaction data-points as well!

### Continous Data

Often, continous data should be explored with specific domain knowledge, common transformations include
1. Square Rooting
2. Log
3. Squaring
4. Interaction (V1 * V2)
5. + / - *
6. Scaling (Normalize & Standardize)

### Missing Data

Missing data can be handled in many different ways, each with pros and cons
1. Replace with 0
2. Drop rows
3. Drop column
4. Replace with Mean, Median, Mode
5. Build a model
6. Factorization
