Internal Evaluation - How do we feel about the groups we are in?  Let's pick a representative for the group, does she/he represent the group?

External Evaluation - Let's compare our representatives?  How different are they? Are they more different than the individuals in the group or less?

Let's organize ourselves based on this evaluation- representatives more similar move together!
